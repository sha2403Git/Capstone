```python
def stringToInt(var):
    try:
        var = np.int(var)
        return var
    except Exception:
        return np.int(0)
    
fData=pd.read_csv('data/historical_data1_Q12010/historical_data1_Q12010.txt','|',index_col=None, encoding='utf-8',low_memory=False
                 ,converters={'CREDIT SCORE':stringToInt})
#fData.info()
#fData.head()
## <640, 640-680,680-720,720-850
fData1=fData['CREDIT SCORE']
fData1.plot.hist(xlim=[300,900],bins=50)
```
